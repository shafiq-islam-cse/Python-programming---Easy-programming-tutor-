```python
#2 For Loop

fruits = ["apple", "banana", "cherry", "Nuts", "100 Oranges"]
for x in fruits:
  print(x)
  
```

```python
#fruits = ["apple", "banana", "cherry"]
fruits = ["apple" "cherry", "Nuts", "100 Oranges", "banana", "Nuts",]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```
