```python
"""
Example of using if and else condition in python while loop
Print a message once the condition is false:
"""
i = 1
while i < 15:
  print(i)
  i += 2# i = i+2
else:
  print("i is no longer less than 6")
```

```python
#2. Nested while loop in pyhon programming
#!/usr/bin/python

i = 1
j = 5
while i < 4:
    while j < 8:
        print(i, ",", j)
        j = j + 1
        i = i + 1
print ("Good bye!")
```
